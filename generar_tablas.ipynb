## Dimensión departamento

Lo que se quiere hacer es hallar el PIB per cápita de cada departamento.

```python
economia_copy = economia.copy()
economia_copy = economia_copy[economia_copy['Indicador'] == 'PIB']

economia_copy['Dato Numérico'] = (economia_copy['Dato Numérico'].apply(lambda x: x.replace('.', '').replace(',', '.'))).astype('float64')
dimension_departamento = economia_copy.groupby(by = ['Código Departamento', 'Departamento']).sum()
dimension_departamento.rename(columns={'Dato Numérico': 'PIB'}, inplace=True)

for col in ('Código Entidad', 'Dato Cualitativo', 'Año', 'Mes'):
  del dimension_departamento[col]

dimension_departamento.reset_index(inplace=True)

dimension_departamento.drop(0, inplace=True)
dimension_departamento.rename(columns={'Código Departamento': 'id_departamento', 'Departamento': 'departamento'}, inplace=True)
dimension_departamento.to_csv('dimensiones/dimension_departamento.csv', sep=';', index=False)
```

## Dimensión municipio

Es perentorio revisar que se estén agrupando los datos como corresponde.

```python
demografia_personas = demografia[demografia['Unidad de Medida'] == 'Personas']
dimension_municipio = demografia_personas.groupby(by = ['Código Departamento', 'Código Entidad', 'Entidad']).sum()
del dimension_municipio['Año']

dimension_municipio.reset_index(inplace=True)
dimension_municipio.rename(columns = {
    'Código Departamento': 'id_departamento',
    'Código Entidad': 'id_municipio',
    'Entidad': 'nombre',
    'Dato Numérico': 'poblacion'
}, inplace=True)

for id_departamento in dimension_municipio['id_departamento'].unique():
    dimension_municipio.drop(dimension_municipio[dimension_municipio['id_municipio'] == (id_departamento*1000)].index, inplace=True)
    
dimension_municipio.drop(0, inplace=True)
dimension_municipio
```

```python
dimension_municipio.to_csv('dimensiones/dimension_municipio.csv', index=False)
```
