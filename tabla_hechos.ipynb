```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
tabla_hechos = pd.read_csv('datos/data_2010_2017.csv', sep=',')
```

```python
tabla_hechos
```

```python
nuevos_nombres = {
    'codigo_depto': 'id_departamento',
    'codigo_muni': 'id_municipio',
    'codigo_evento': 'id_tipo_muerte',
    'evento': 'tipo_muerte',
    'edad': 'grupos_edad',
    'anio': 'fecha'
}

tabla_hechos.rename(columns=nuevos_nombres, inplace=True)
```

```python
for col in tabla_hechos.columns:
    tabla_hechos = tabla_hechos[tabla_hechos[col] != 'TOTAL']
```

```python
tabla_hechos
```

```python
tabla_hechos.drop(['nombre_depto', 'nombre_muni'], axis='columns', inplace=True)
tabla_hechos
```

```python
for col in ('tipo_muerte', 'grupos_edad', 'sexo'):
    tabla_hechos[col] = tabla_hechos[col].astype('category')
```

```python
tabla_hechos = tabla_hechos[tabla_hechos['id_departamento'] != 'Extranjeros']
```

```python
for col in ('id_departamento', 'id_municipio', 'num_casos', 'fecha'):
    tabla_hechos[col] = tabla_hechos[col].astype('int32')
```

```python
for col in ('fecha', 'grupos_edad', 'sexo'):
    tabla_hechos['id_' + col] = LabelEncoder().fit_transform(tabla_hechos[col])
    
tabla_hechos
```

```python
tabla_hechos
```

```python
dimension_grupos_edad = tabla_hechos[['id_grupos_edad', 'grupos_edad']].copy().drop_duplicates()
```

```python
dimension_grupos_edad
```

```python
dimension_grupos_edad.to_csv('dimensiones/dimension_grupos_edad.csv', sep=';', index=False)
```

```python
dimension_sexo = tabla_hechos[['id_sexo', 'sexo']].copy().drop_duplicates()
```

```python
dimension_sexo
```

```python
dimension_sexo.to_csv('dimensiones/dimension_sexo.csv', sep=';', index=False)
```

```python
dimension_tipo_muerte = tabla_hechos[['id_tipo_muerte', 'tipo_muerte']].copy().drop_duplicates()
```

```python
dimension_tipo_muerte
```

```python
dimension_tipo_muerte.to_csv('dimensiones/dimension_tipo_muerte.csv', sep=';', index=False)
```

```python
dimension_fecha = tabla_hechos[['id_fecha', 'fecha']].copy().drop_duplicates()
```

```python
dimension_fecha
```

```python
dimension_fecha.to_csv('dimensiones/dimension_fecha.csv', sep=';', index=False)
```
