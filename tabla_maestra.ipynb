# Creación tabla maestra

```python
dimensiones = (
    dimension_departamento,
    dimension_fecha,
    dimension_grupos_edad,
    dimension_municipio,
    dimension_sexo,
    dimension_tipo_muerte
)
```

```python
intermedia = tabla_hechos.copy()

for dimension in dimensiones:
    index_name = dimension.index.names[0]
    intermedia = pd.merge(intermedia, dimension, on=index_name, how='inner')
```
